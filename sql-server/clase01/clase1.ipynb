# ¿Qué es SQL Server?

Es un sistema de gestión de bases de datos relacionales, esta diseñada para proporcionar un entorno robusto, escalable y seguro para la gestión de los datos. 

## Nuevas funciones de SQL Server 2022

- Integración con el servicio en la nube de Azure, permitiendo realizar copias de seguridad con Azure Blob Storage y aprovechar Azure Synapse para el análisis en tiempo real.
- Inteligencia artificial y machine learning, que permiten implementar modelos para mejorar la toma de decisiones.
- Mejora de rendimiento en la optimización de consultas y la gestión de índices.
- Seguridad mejorada, como el cifrado de datos y una auditoría optimizada.
- Recuperación ante desastres automática con failover y copias de seguridad en Azure, permitiendo mantener la disponibilidad de los datos en caso de fallos.

## Componentes

- Motor de base de datos: Es el núcleo que maneja las operaciones de almacenamiento y recuperación de datos.
- SQl Server Management Studio (SSMS): Herramienta para la administración de base de datos, ejecución de consultas y mantenimiento.
- SQL Server Data Tools (SSDT): Herramienta para diseñar y gestionar base de datos.
- Reporting Services (SSRS): Herramienta para la creación y gestión de reportes.
- Integration Services (SSIS): Herramienta para la integración de datos y flujos de trabajo de ETL.
- Analysis Services (SSAS): Herramienta para el análisis de datos y creación de cubos OLAP.

## Creación de una base de datos

- La base de datos debe diseñarse para cubrir un horizonte de al menos 5 años de datos. Si no se considera este periodo, puede fragmentarse, causando demoras en el rendimiento.
Ejemplo de años a considerar: 2020, 2021, 2022, 2023, 2024 y más allá, como "2025".

- Es crucial prever un crecimiento regular de la base de datos para facilitar el mantenimiento a lo largo del tiempo. Esto evitará problemas de capacidad y optimizará su rendimiento.

- SQL Server organiza la base de datos en páginas, cada una ocupando 8 KB en el disco.
Las páginas se agrupan en extensiones de 64 KB, que contienen 8 páginas consecutivas, funcionando como bloques de almacenamiento optimizados.

- El log debe ocupar entre el 25% y el 30% del tamaño de la base de datos original. Un log excesivamente grande puede impactar negativamente el rendimiento y complicar el mantenimiento.

El log es el archivo .ldf  y el dat donde se maneja la data es mdf

Recordar que los archivos deben estar en otras unidades y no en la unidad C es caso de ataques.



## ¿Qué es SQL?

Es un lenguaje declarativo que yo especifico que operación voy a realizar y el sistema de gestión de base de datos (DBMS) se encarga de ejecutar las operaciones.

## Tipos de Comando SQL

| **Tipo** | **Propósito**                     | **Ejemplos de Comandos**         |
|----------|-----------------------------------|----------------------------------|
| **DDL**  | Definir estructura               | `CREATE`, `ALTER`, `DROP`, `TRUNCATE`, `RENAME` |
| **DML**  | Manipular datos                  | `SELECT`, `INSERT`, `UPDATE`, `DELETE` |
| **DCL**  | Controlar permisos               | `GRANT`, `REVOKE`               |
| **TCL**  | Gestionar transacciones          | `COMMIT`, `ROLLBACK`, `SAVEPOINT`, `SET TRANSACTION` |
| **DQL**  | Consultar datos                  | `SELECT`                        |
