## Base de datos Master

- Contiene información sobre todas las demás bases de datos en el servidor, detalles del servicio, usuarios.

## Base de datos Model

- Proporciona un modelo que va a servir como plantilla para las nuevas bases de datos que se crean con las mismas caracteristicas que la model en tamaño, etc.
- Si yo le pongo los parametros va a ir cambiando.

Para crear se debe tener 5 años anteriores para ver el tamaño que va a tener
2019, 2020, 2021, 2022, 2023, 2024.

La base de datos se contruye por páginas como si fuera un libro y cada página ocupa 8 KByte en el disco.
El log debe ser la tercera parte o el 25% de la base de datos original no debe pasar de ese tamaño.





## Base de datos msdb

- Se utiliza para almacenar información sobre trabajos programados, alertas y eventos realcionados con la administración y el mantenimiento del servidor. Se realiza tareas automatizadas.

## Base de datos tempdb

- Cuando se crear una tabla temporal no se crean en la base del usuario si no una temporal de la base de datos del sistema.

## Esquema

- Es una forma de organizar y agrupar objetos de la base de datos y gestionar mejor la seguridad de acceso a los datos.
